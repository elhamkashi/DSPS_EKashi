```python
earthquakes.columns  #Listing All Column Names
```

```python
earthquakes = earthquakes[['#YYY/MM/DD', 'HH:mm:SS.ss', "MAG"]]  #Selecting a Subset of Columns, only the columns relevant to the current analysis and discards the rest
```

```python
#earthquakes[['#YYY/MM/DD', 'HH:mm:SS.ss', 'ET', 'GT', 'MAG']]
```

```python
earthquakes.dtypes
```

```python
earthquakes.rename({"#YYY/MM/DD": "date", "HH:mm:SS.ss": "time", "MAG":"mag" },
                   axis=1, inplace=True)
# change the column names to simpler names , {old_name: new_name, ...}
#axis=0 for raws, inplace=True=apply this change directly to the original,
```

```python
earthquakes.rename({"#YYY/MM/DD": "date", "HH:mm:SS.ss": "time",  "MAG": "mag"}, axis=1)   #with  inplace=True, it doesnot show the results
```
