```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, t

# 1. Generate a range of x values to plot the distributions over
x = np.linspace(-5, 15, 1000)  # From -5 to 15 to capture all centers

# 2. Create the figure and axis
plt.figure(figsize=(12, 6))

# 3. Generate and plot the four distributions

# Standard Normal (mean=0, variance=1)
normal_0 = norm.pdf(x, loc=0, scale=1) # scale is standard deviation (sqrt(variance))
plt.plot(x, normal_0, label='Normal(μ=0, σ²=1)', color='blue', linestyle='-')

# Standard Normal (mean=10, variance=1)
normal_10 = norm.pdf(x, loc=10, scale=1)
plt.plot(x, normal_10, label='Normal(μ=10, σ²=1)', color='red', linestyle='-')

# Student's t (df=2, center=0)
t_0 = t.pdf(x, df=2, loc=0)
plt.plot(x, t_0, label="Student's t (df=2, center=0)", color='green', linestyle='--')

# Student's t (df=2, center=10)
t_10 = t.pdf(x, df=2, loc=10)
plt.plot(x, t_10, label="Student's t (df=2, center=10)", color='orange', linestyle='--')

# 4. Add labels, title, legend, and grid
plt.title("Comparing Normal and Student's t Priors", fontsize=14)
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True, alpha=0.3)

# 5. Show the plot
plt.tight_layout()
plt.show()
```
